Transform to perform some manipulation of the data and make it suitable of training, as data does not always come in its final processed format.

### Torchvision models:
- **transform**: to modify the features
- **target_transform**: to modify the labels

**Example:**

FashionMNIST: labels are integers, and we need them as one-hot encoded tensors. images are in PIL format, and we need features as normalized tensors

```python
import torch
from torch import nn
from torchvision.transforms import ToTensor, Lambda
from torchvision.datasets import FashionMNIST
from torch.utils.data import DataLoader, Dataset
```

```python
torch.zeros(10, dtype=torch.float).scatter(0, torch.tensor(4), value=1)
```

```python
train_dataset = FashionMNIST(
    root="data",
    train=True,
    download=True,
    transform=ToTensor(),
    target_transform=Lambda(lambda y: torch.zeros(10, dtype=torch.float).scatter_(0, torch.tensor(y), value=1))
)

test_dataset = FashionMNIST(
    root="data",
    train=False,
    transform=ToTensor(),
    download=True,
    target_transform=Lambda(lambda y: torch.zeros(10, dtype=torch.float).scatter_(0, torch.tensor(y), value=1))
)

train_dataset, test_dataset
```

```python
for X, y in train_dataset:
    print(X, y)
    print(X.shape, y.shape)
    break
```

```python
train_dataloader = DataLoader(train_dataset, batch_size=64, shuffle=True)

test_dataloader = DataLoader(test_dataset, batch_size=64, shuffle=True)

train_dataloader, test_dataloader
```

```python
for X, y in train_dataloader:
    print(X.shape, y.shape)
    break
```

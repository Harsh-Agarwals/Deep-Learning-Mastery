```python
from langchain_community.document_loaders import ArxivLoader

loader = ArxivLoader(query="1706.03762")
doc = loader.load()
doc
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

text_splitter=RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=50)
# It usually splits by "\n\n", "\n", etc
final_doc = text_splitter.split_documents(doc)
final_doc
```

```python
from langchain_community.document_loaders import TextLoader

loader = TextLoader("doc.txt")
doc = loader.load()
doc
```

```python
with open("doc.txt") as f:
    speech=f.read()

speech
```

```python
ts = RecursiveCharacterTextSplitter(chunk_size=30, chunk_overlap=10)
split=ts.split_text(speech)
split, type(split), type(split[0])
```

```python
split2=ts.create_documents([speech])
split2, type(split2), type(split2[0])
```

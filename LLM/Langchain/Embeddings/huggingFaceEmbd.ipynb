```python
from langchain_community.embeddings import HuggingFaceEmbeddings

model_name="sentence-transformers/all-mpnet-base-v2"
model_kwargs={'device': 'cpu'}
encode_kwargs={'normalise_embeddings': False}
embd = HuggingFaceEmbeddings(model_name=model_name, model_kwargs=model_kwargs, encode_kwargs=encode_kwargs)
```

```python
embd
```

```python
emb = embd.embed_query("Napoletana is the best pizza variety, from naples.")
emb, len(emb)
```

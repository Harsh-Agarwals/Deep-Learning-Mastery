```python
from transformers import pipeline
```

```python
tts = pipeline(task="text-to-speech", model="suno/bark", device="cuda")
tts("Hello, I am Harsh and I live in Mumbai. India")
```

```python
%%time

output = tts("Hello, I am Harsh and I live in Mumbai, India")
print(output['audio'].shape)

from IPython.display import Audio
Audio(output['audio'], rate=output['sampling_rate'])
```

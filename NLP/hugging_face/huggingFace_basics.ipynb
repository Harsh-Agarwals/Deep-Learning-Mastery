```python
from transformers import pipeline
```

```python
# Question-answering

oracle = pipeline(model="deepset/roberta-base-squad2")
oracle(question="Where do I live?", context="My name is Wolfgang and I live in Berlin")
```

### Sentiment Analysis

```python
sent = pipeline("sentiment-analysis")
sent("Mumbai is an awful terrible city")
```

```python
sent(["Mumbai is an awful terrible city", "I dont understand how people like mumbai", "Traffic, congestion, humidity and pollution are the major issues of mumbai"])
```

```python
txtclf = pipeline(task="text-classification", model="facebook/bart-large-mnli")
txtclf(["Mumbai is an awful terrible city", "I dont understand how people like mumbai", "Traffic, congestion, humidity and pollution are the major issues of mumbai"])
```

```python
txtclf = pipeline(task="sentiment-analysis", model="SamLowe/roberta-base-go_emotions")
txtclf(["Mumbai is an awful terrible city", "I dont understand how people like mumbai", "Traffic, congestion, humidity and pollution are the major issues of mumbai"])
```

### Question Answering

```python
qna = pipeline(task="question-answering")
question = "What I wanted to shift to?"
context = "I am an ML developer looking to pivot to GenAI."

qna(question=question, context=context)
```

```python
qa_model = pipeline("question-answering")
question = "What is my job?"
context = "I am developing AI models with Python."
qa_model(question = question, context = context)
```

### Tokenizers

```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification, DistilBertForSequenceClassification
```

```python
model_name2 = "nlptown/bert-base-multilingual-uncased-sentiment"
model2 = AutoModelForSequenceClassification.from_pretrained(model_name2)
tokenizer2 = AutoTokenizer.from_pretrained(model_name2)

model2, tokenizer2
```

```python
clf = pipeline(task="sentiment-analysis", model=model2, tokenizer=tokenizer2)
clf(["San Francisco is the best for any techie to be at this moment."])
```

```python
tokenizer2.tokenize("San Francisco is the best for any techie to be at this moment.")
```

```python
tokenizer2.convert_tokens_to_ids(tokenizer2.tokenize("San Francisco is the best for any techie to be at this moment."))
```

```python
tokenizer2("San Francisco is the best for any techie to be at this moment.")
```

```python
tokenizer2.decode(tokenizer2.convert_tokens_to_ids(tokenizer2.tokenize("San Francisco is the best for any techie to be at this moment.")))
```
